```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('dados.csv', sep=';')
```

```python
df.shape
```

```python
df.head()
```

```python
# Checamos se tem algum valor vazio
df.isnull().sum()
```

## Histograma do PIB

```python
sns.histplot(data=df, x='PIB', kde=True).set_title('PIB');
plt.show()
# 
```

- nota-se uma assimetria à direita (media > mediana)
- nota-se um outlier (Aprox. 200.000)

```python
df['PIB'].describe()
```

- outlier confirmado 223848.22

## Boxplot do PIB

```python
sns.boxplot(data=df['PIB']).set_title('PIB')
# É possível observar alguns pontos intermediários
```

## Histograma do Valor Empenho

```python
sns.histplot(df['VALOREMPENHO'], kde=True).set_title('Valor Empenho');
plt.xlabel('Valor Empenho')
```

```python
df['VALOREMPENHO'].describe()
```

- nota-se uma assimetria à direita (media > mediana)
- outliers nessa variavel estão presentes também

## Boxplot do Valor Empenho

```python
sns.boxplot(df['VALOREMPENHO']);
```

## Municipios com maiores PIB

```python
top_municipios = df.sort_values('PIB', ascending=False).head(10)
sns.barplot(x='PIB', y='MUNICIPIO', data=top_municipios, hue='MUNICIPIO', palette='viridis');
```

- Outlier maior é de TRIUNFO
- "Outliers" menores indentificados
